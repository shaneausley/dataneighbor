```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, uniform, expon, poisson, bernoulli, binom, t, chi2, lognorm, gamma, beta, nbinom

# Setting up the figure for a 4x3 grid
fig, axes = plt.subplots(4, 3, figsize=(18, 16))
axes = axes.flatten()

# Distributions
distributions = [
    ("Normal", norm.rvs(size=1000)),
    ("Uniform", uniform.rvs(size=1000)),
    ("Exponential", expon.rvs(size=1000)),
    ("Poisson", poisson.rvs(mu=3, size=1000)),
    ("Bernoulli", bernoulli.rvs(p=0.5, size=1000)),
    ("Binomial", binom.rvs(n=10, p=0.5, size=1000)),
    ("t-Distribution", t.rvs(df=10, size=1000)),
    ("Chi-Square", chi2.rvs(df=4, size=1000)),
    ("Log-Normal", lognorm.rvs(s=0.954, size=1000)),
    ("Gamma", gamma.rvs(a=2, size=1000)),
    ("Beta", beta.rvs(a=2, b=5, size=1000)),
    ("Negative Binomial", nbinom.rvs(n=5, p=0.5, size=1000)),
]

# Plot each distribution
for ax, (name, data) in zip(axes, distributions):
    sns.histplot(data, bins=30, kde=True, ax=ax)
    ax.set_title(name)
    ax.set_xlabel("")
    ax.set_ylabel("")

plt.tight_layout()
plt.show()
```
